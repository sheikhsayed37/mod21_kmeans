````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

````

````python
df=load_breast_cancer(as_frame=True).frame
df.shape
````

````python
df
````

````python
x=df.drop('target',axis=1)
y=df['target']
````

````python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
````

````python
# full dastaset e pca use kore

np.set_printoptions(suppress=True)   #  point value shpw korar jonno

pca_full=PCA(n_components=None)

x_pca_full=pca_full.fit_transform(x_scaled)

explained_part=pca_full.explained_variance_ratio_

explained_part


````

````python
cumualtive_variance=np.cumsum(explained_part)

cumualtive_variance      # j col e ashe tar ager ter sathe jog korse
````

````python
plt.plot(cumualtive_variance , marker='o')
plt.xlabel("Number of components")
plt.ylabel("Cumulative variance")
plt.grid()
plt.show()
````

````python
# comparism with logistic Regression
x_train,x_test,y_train,y_test=train_test_split(
   x,y,test_size=0.2,random_state=42,stratify=y)


# Pipeline WITHOUT PCA: Scale -> Logistic Regression
pipe_no_pca=Pipeline([
    ("scaler",StandardScaler()),
    ("clf",LogisticRegression(max_iter=400))
])

pipe_no_pca.fit(x_train,y_train)

y_pred_no_pca = pipe_no_pca.predict(x_test)

acc_no_pca = accuracy_score(y_test,y_pred_no_pca)

print("Accuracy (No PCA):", acc_no_pca)
````

````python
# PCA FIRST 10 COL
# LOGODTIC REGRESSION e pca korle ki hoi seta judge kora mane accuracy change
# kmnne hocce ta measurement kora

pipe_with_pca=Pipeline([
    ("scaler",StandardScaler()),
    ("pca",PCA(n_components=10)),
    ("clf",LogisticRegression(max_iter=400))
])


pipe_with_pca.fit(x_train ,y_train)
y_pred_with_pca=pipe_with_pca.predict(x_test)

acc_with_pca=accuracy_score(y_test,y_pred_with_pca)


print("Accuracy (With PCA):",acc_with_pca)


````



**PCA-এর "First Component (PC1)" মানে dataset-এর প্রথম column নয়।**

ধরো তোমার dataset:

| Radius | Texture | Area |
| -----: | ------: | ---: |
|     10 |      15 |  100 |
|     12 |      16 |  120 |
|     15 |      18 |  150 |

এখানে **প্রথম column = Radius**।

কিন্তু PCA করার পর dataset হয়ে যেতে পারে:

| PC1 | PC2 | PC3 |
| --: | --: | --: |
| ... | ... | ... |
| ... | ... | ... |

এখানে **PC1 = Radius নয়**।

বরং,

```text
PC1 = (0.45 × Radius)
    + (0.30 × Texture)
    + (0.84 × Area)
```

অর্থাৎ **PC1 হলো সব feature-এর একটি weighted combination (linear combination)।**

---

### তাহলে "First Component" কেন বলে?

কারণ PCA সব possible direction-এর মধ্যে এমন একটি নতুন direction খুঁজে বের করে যেখানে **variance সবচেয়ে বেশি**।

সেই direction-কে **Principal Component 1 (PC1)** বলে।

এরপর দ্বিতীয় সর্বোচ্চ variance-এর direction = **PC2**।

---

### মনে রাখো

❌ **PC1 ≠ প্রথম column**

✅ **PC1 = সব feature-এর combination, যেখানে variance সবচেয়ে বেশি।**

**এক লাইনে:**

> **Dataset-এর first column আর PCA-এর first component সম্পূর্ণ আলাদা জিনিস। First component হলো PCA তৈরি করা একটি নতুন feature, যা original feature-গুলোর combination।**


এটাই PCA-এর সবচেয়ে গুরুত্বপূর্ণ ধারণাগুলোর একটি।

**Variance সবসময় প্রথমটাতে (PC1) বেশি থাকে না, বরং PCA এমনভাবে তৈরি করা হয় যে প্রথম component (PC1)-এ সবচেয়ে বেশি variance থাকবে।**

ধরো তোমার data এমন:

```text
Height   Weight
160       50
165       55
170       60
175       65
180       70
```

দেখো, Height বাড়লে Weight-ও বাড়ছে।

Graph আঁকলে data এমন হবে:

```text
Weight
70          ●
65        ●
60      ●
55    ●
50  ●
   --------------------> Height
```

Data **তির্যক (diagonal)** দিকে ছড়িয়ে আছে।

PCA দুইটা নতুন direction বানায়:

```text
        PC2
         ↑
         |
      ●
    ●
  ●
●
--------------------→ PC1
```

* **PC1** data যেদিকে সবচেয়ে বেশি ছড়িয়েছে, সেই direction।
* **PC2** তার সাথে ৯০° কোণে থাকে।

তাই,

* **PC1-এর variance সবচেয়ে বেশি।**
* **PC2-এর variance PC1-এর চেয়ে কম।**

### কেন?

কারণ PCA-এর কাজই হলো:

> **প্রথম component এমনভাবে নির্বাচন করা যাতে সেটিতে সর্বোচ্চ variance থাকে।**

এরপর,

> **দ্বিতীয় component এমনভাবে নির্বাচন করা হয় যাতে এটি PC1-এর লম্ব (orthogonal) হয় এবং বাকি থাকা সর্বোচ্চ variance ধরে।**

এভাবে PC3, PC4... তৈরি হয়।

তাই **PC1-এ সবসময় সবচেয়ে বেশি information/variance থাকে**, কারণ PCA ইচ্ছাকৃতভাবে সেটাকেই প্রথমে বেছে নেয়।

### এক লাইনে মনে রাখো

> **PC1-এর variance সবচেয়ে বেশি, কারণ PCA-এর algorithm প্রথমেই এমন direction খুঁজে বের করে যেখানে data সবচেয়ে বেশি spread (variance) থাকে।**


### **PCA (Principal Component Analysis) – সংক্ষিপ্ত Summary**

* **PCA** একটি **Dimensionality Reduction** technique।
* এটি অনেক **feature** থেকে কম সংখ্যক **Principal Component (PC)** তৈরি করে।
* এই নতুন component-গুলো original feature-এর **linear combination**।
* **PC1** সবচেয়ে বেশি information (variance) ধরে রাখে, **PC2** দ্বিতীয় সর্বোচ্চ, এভাবে চলতে থাকে।
* PCA ব্যবহার করার আগে সাধারণত **StandardScaler** ব্যবহার করা হয়, যাতে সব feature একই scale-এ থাকে।
* `explained_variance_ratio_` প্রতিটি component কত শতাংশ information ধরে রেখেছে তা দেখায়।
* `np.cumsum(explained_variance_ratio_)` cumulative variance দেখায়, অর্থাৎ প্রথম কয়েকটি component মিলে মোট কত information রাখা হয়েছে।
* `n_components=10` দিলে ৩০টি feature-এর পরিবর্তে শুধু ১০টি principal component রাখা হয়।
* PCA ব্যবহারের সুবিধা:

  * Feature সংখ্যা কমে।
  * Training দ্রুত হয়।
  * Memory কম লাগে।
  * Noise ও Overfitting কমতে পারে।
* শেষে **PCA ছাড়া** এবং **PCA সহ** Logistic Regression-এর **accuracy** তুলনা করে দেখা হয়, PCA ব্যবহার করলে performance কতটা পরিবর্তন হয়েছে।

### PCA-তে Variance কেন দরকার?

PCA মনে করে:

> **যে direction-এ data সবচেয়ে বেশি ছড়িয়ে (variance বেশি), সেই direction-এ সবচেয়ে বেশি information থাকে।**

তাই,

* **PC1** → Highest Variance
* **PC2** → Second Highest Variance
* **PC3** → Third Highest Variance

### এক লাইনে মনে রাখো

> **Variance = Data কতটা spread (ছড়িয়ে) আছে তার পরিমাপ। PCA বেশি variance-যুক্ত direction-কে আগে নির্বাচন করে।**
