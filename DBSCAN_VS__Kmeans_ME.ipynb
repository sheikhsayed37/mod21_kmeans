````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import  StandardScaler


from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
````

````python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=300, noise=0.06, random_state=42)
X.shape
````

````python
X
````

````python
X, y = make_moons(n_samples=300, noise=0.06, random_state=42)
plt.scatter(X[:,0] , X[:,1])
plt.title('Original Moons Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
````

````python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
````

````python
# scaling korar por plot
plt.scatter(X_scaled[:,0] , X_scaled[:,1])
plt.title('Scaled Moons Dataset')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.show()
````

## 3. Density-Based Spatial Clustering (DBSCAN)
DBSCAN groups points that are closely packed together. It is particularly effective for datasets with complex shapes like our moons, as it doesn't assume clusters are spherical. It also identifies points in low-density regions as outliers (noise).

````python
dbscan=DBSCAN(eps = 0.3 , min_samples =3)

dbscan.fit(X_scaled)

label_dbscan=dbscan.labels_
label_dbscan


# label_dbscan হলো প্রতিটি data point কোন cluster-এ আছে, তার label।

````

label_dbscan হলো প্রতিটি data point কোন cluster-এ আছে, তার label।
array([0, 0, 1, 1, 0])
| Data Point | Cluster Label | অর্থ      |

| 2nd point  | 0             | Cluster 0 |
| 3rd point  | 1             | Cluster 1 |
| 4th point  | 1             | Cluster 1 |
| 5th point  | 0             | Cluster 0 |


````python
sns.scatterplot(x = X_scaled[: ,0] , y =X_scaled[: ,1] , hue = label_dbscan ,palette = 'viridis' )
plt.title('DBSCAN Clustering Results')
plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.legend(title='Cluster')
plt.show()
````

````python
# kemans vs dbscan

kmeans=KMeans(
    n_clusters=2,
    init="k-means++",
    max_iter=300,
    random_state=42)


kmeans.fit(X_scaled)
label_kmeans=kmeans.labels_
label_kmeans
````

````python

sns.scatterplot(x = X_scaled[: ,0] , y =X_scaled[: ,1] , hue = label_kmeans ,palette = 'viridis' )
plt.title('K-Means Clustering Results')
plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.legend(title='Cluster')
plt.show()
````

# **DBSCAN (Density-Based Spatial Clustering of Applications with Noise) – Viva Notes**

## 1. What is DBSCAN?

**Definition:**
DBSCAN হলো একটি **Density-Based Unsupervised Clustering Algorithm**। এটি data-এর density দেখে cluster তৈরি করে এবং noise (outlier) আলাদা করতে পারে।

---

## 2. কেন DBSCAN ব্যবহার করি?

* K আগে থেকে বলতে হয় না।
* Irregular (non-linear) shape-এর cluster খুঁজে বের করতে পারে।
* Noise/Outlier detect করতে পারে।

---

## 3. Parameters

### `eps`

একটি point-এর চারপাশের radius (neighborhood)।

* `eps` ছোট হলে → বেশি noise হতে পারে।
* `eps` বড় হলে → আলাদা cluster একসাথে মিশে যেতে পারে।

---

### `min_samples`

একটি point-কে **Core Point** বলার জন্য minimum কতটি point লাগবে।

উদাহরণ:

```python
DBSCAN(eps=0.3, min_samples=3)
```

মানে radius-এর ভিতরে কমপক্ষে ৩টি point থাকতে হবে।

---

## 4. Types of Points

### Core Point

যার চারপাশে (`eps` radius-এর মধ্যে) কমপক্ষে `min_samples` point থাকে।

---

### Border Point

নিজে Core Point নয়, কিন্তু কোনো Core Point-এর কাছে থাকে।

---

### Noise Point

কোনো cluster-এর অংশ নয়।

DBSCAN-এ Noise-এর label হয়:

```python
-1
```

---

## 5. কেন StandardScaler ব্যবহার করি?

DBSCAN distance ব্যবহার করে।

যদি feature-এর scale আলাদা হয়, তাহলে distance ভুল হতে পারে।

তাই আগে StandardScaler ব্যবহার করা হয়।

---

## 6. Graph কেন করি?

### Original Graph

Dataset-এর আসল shape দেখার জন্য।

---

### Scaled Graph

Scaling-এর পরে shape ঠিক আছে কিনা দেখার জন্য।

---

### DBSCAN Graph

* কতগুলো cluster হয়েছে।
* Noise আছে কিনা।
* Algorithm ঠিকমতো cluster করেছে কিনা।

---

## 7. Moon Dataset-এ DBSCAN কেন ভালো?

কারণ Moon dataset-এর cluster **বাঁকা (non-linear)**।

DBSCAN irregular shape-এর cluster সহজে detect করতে পারে।

---

## 8. Evaluation

DBSCAN-এর জন্য সাধারণত:

* Graph
* Silhouette Score
* Davies-Bouldin Index

ব্যবহার করা হয়।

**Accuracy, Precision, Recall, F1-score ব্যবহার করা হয় না**, কারণ এটি Unsupervised Learning।

# Viva Questions

### Q1. What is DBSCAN?

**Answer:**
DBSCAN is a density-based unsupervised clustering algorithm that groups data points based on density and detects noise.

---

### Q2. What does `eps` mean?

**Answer:**
`eps` is the radius used to find neighboring points.

---

### Q3. What is `min_samples`?

**Answer:**
It is the minimum number of points required to form a core point.

---

### Q4. What does label `-1` mean?

**Answer:**
`-1` indicates a noise (outlier) point.

---

### Q5. Why do we use StandardScaler before DBSCAN?

**Answer:**
Because DBSCAN uses distance calculations, scaling ensures all features contribute equally.

---

### Q6. Why is DBSCAN better than K-Means for the Moon dataset?

**Answer:**
Because DBSCAN can detect non-linear and irregular-shaped clusters, while K-Means works best with spherical clusters.

---

### Q7. Does DBSCAN need the number of clusters beforehand?

**Answer:**
No. DBSCAN automatically determines the number of clusters based on data density.


| **K-Means**                                  | **DBSCAN**                                  |
| -------------------------------------------- | ------------------------------------------- |
| **Centroid-based** algorithm                 | **Density-based** algorithm                 |
| আগে থেকে **K (number of clusters)** দিতে হয় | K আগে থেকে দিতে হয় না                      |
| গোলাকার (spherical) cluster-এর জন্য ভালো     | Irregular (non-linear) cluster-এর জন্য ভালো |
| Noise/Outlier আলাদা করতে পারে না             | Noise/Outlier detect করতে পারে (`-1`)       |
| প্রতিটি point কোনো না কোনো cluster-এ থাকে    | কিছু point noise হিসেবে থাকতে পারে          |
| `n_clusters` প্রধান parameter                | `eps` এবং `min_samples` প্রধান parameter    |
| Inertia ব্যবহার করে evaluation করা যায়      | Inertia ব্যবহার করা যায় না                 |
| Elbow Method দিয়ে K নির্বাচন করা হয়        | Elbow Method লাগে না                        |
